```python
dict_og = {'id':[],'og':[]}
for index, i in df.iterrows():
    try:
        ogs = i['eggNOG OGs'].split(',')
    
        for j in ogs:
            n = j.split('@')
            if int(n[1]) == 1:
                best_og = n[0]
    except:
        best_og = 'NaN'
    dict_og['id'].append(i['#query_name'])
    dict_og['og'].append(best_og)
```

```python
dict_ = {i:[] for i in dict_og['og']}

for index, item in enumerate(dict_og['id']): 
    dict_[dict_og['og'][index]].append(item)
```

```python
from Bio import SeqIO
file = 'og_2C38W.ali'  

dict_topt = {'id':[] , 'ot':[], 'seq':[]}

for i, rec in enumerate(SeqIO.parse(file,'fasta')):
    if rec.id in dict_['2C38W']:
        dict_topt['id'].append(rec.id)
        dict_topt['seq'].append(str(rec.seq))
        dict_topt['ot'].append(rec.description.split()[-1])

#f = open('og_2C38W.ali', 'w')
#for i, seq, ot in zip(dict_topt['id'], dict_topt['seq'], dict_topt['ot']):
#    print(i)
#    f.write('>{} {}\n{}\n'.format(i, ot, seq))
```

```python
count = 0
dist_avg = 0
dist_len_avg = 0
for i in range(len(dict_topt['id'])-1):
    for j in range(i,len(dict_topt['id'])): 
        dist_avg += lev.distance(dict_topt['seq'][i],dict_topt['seq'][j])
        count += 1
print(dist_avg)
print(count)
print((dist_avg-dist_len_avg)/count)
```
